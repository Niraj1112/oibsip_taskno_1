IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

LOADING THE DATASET

```python
#Loading the dataset#
iris = pd.read_csv('/content/Iris.csv')

```

```python
# Rename the complex columns name
iris= iris.rename(columns={'SepalLengthCm':'Sepal_Length',
                           'SepalWidthCm':'Sepal_Width',
                           'PetalLengthCm':'Petal_Length',
                           'PetalWidthCm':'Petal_Width'})
```

```python
iris.head()
```

```python
iris.tail()
```

```python
# checking if the data is biased or not
iris ['Species'].value_counts()
```

SEPARATING INPUT COLUMNS AND OUTPUT COLUMNS

```python
# Separate features and target
data=iris.values

# slicing the matrices
X=data[:,0:4]
Y=data[:,5]
```

```python
print(X.shape)
print(X)
```

```python
print(Y.shape)
print(Y)
```

SPLITTING THE DATA INTO TRAINING AND TESTING

```python
#Split data to train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2)
```

MODEL 1: KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn=KNeighborsClassifier(n_neighbors=3)
model_knn.fit(X_train,y_train)
```

```python
prediction2 = model_knn.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2))
```

MODEL 2: SVM


```python
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1 = model_svc.predict(X_test)

#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1))
```

MODEL 3: LOGISTIC REGRESSION

```python
# converting categorical variables into numbers
flower_mapping = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris['Species']=iris['Species'].map(flower_mapping)
```

```python
# preparing inputs and outputs
X=iris [['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']].values
y= iris[['Species']].values
```

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X,y)
```

```python
#accuracy
model.score(X,y)
```

```python
#Make predictions for all 150 datasets

expected = y
predicted = model.predict(X)
predicted
```

NEW DATA FOR PREDICTION

```python
# New data for prediction
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])

# Predicting the sizes of the iris flowers
predicted_sizes = model.predict(X_new)

# Output the predicted sizes
print(predicted_sizes)
```
